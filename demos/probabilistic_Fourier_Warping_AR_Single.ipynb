# Demo - Probabilistic Fourier with AR model on coefficients

This file provides an demo of the probabilistic Fourier framework as outline in "Bayesian spectrum estimation of unevenly sampled data" - Qi et al.

$$ {\bf{s}}_t = {\bf{s}}_{t-1} + {\bf{w}}_t $$
$$ {\bf{y}}_t = C^\top {\bf{s}}_t + {\bf{v}}_t $$

## Load packages and functions

```julia
using ForneyLab       # based on last official release
using LinearAlgebra
using Plots
using PyPlot
using ProgressMeter
using Compat
using WAV
using Base64
using FileIO
using HDF5
using FLAC
using DSP
using FFTW
using LAR
using LAR.AR
using LAR.DataAR

include("../functions/auxiliary/buffer.jl")
include("../functions/auxiliary/audioplayer.jl")
include("../functions/auxiliary/workflow.jl")
include("../functions/auxiliary/fourier.jl")
include("../functions/auxiliary/warpedfilter.jl")
;
```

## Settings

```julia
# data loading
sound_file = "../data/obama.wav"   # specify sound location
fs_desired = 8000                  # specify desired sampling frequency

# input buffer parameters
step_size = 1                      # specify step size of buffer
buffer_size = 1                    # specify size of buffer

# warping
nr_taps_warped = 32                 # number of taps of warped filter bank
z_pole = 0.4092                    # pole location of all-pass filter segments (0.5756 ~ approximation Bark scale)
window = rectangularwindow         # specify windowing function from fourier.jl file

# probabilistic fourier parameters 
Ïƒ_observation = 0.001               # specify observation noise std
freq_bins = 15
freq = collect(fs_desired/(2*(freq_bins+1)):fs_desired/(2*(freq_bins+1)):(fs_desired/2-fs_desired/(2*(freq_bins+1))))

# AR modelling
AR_order = 5
iterations = 3
;
```

## Load data

```julia
# load and process speech signal
x_time, fs_time = wavread(sound_file, format="native")
fs_time = Int(fs_time)
x_time = x_time[24*fs_time+1:27*fs_time]
x_time = resample(x_time, fs_desired/fs_time)
x_time = 100*x_time/var(x_time)

# specify time axis
t = collect((1:length(x_time))/fs_desired)

# create audioplayer
audioplayers(x_time, fs=fs_desired)
;
```

## Visualize warped data

```julia
## initialize buffer
buf = BufferData(x_time, t, buffer_size)

# initialize memory vector of taps, Y, with zeros
Y = zeros(nr_taps_warped*2,1)

# initialize memory struct for filter taps and spectrum coefficients
mem_taps = Array{Array{Float64,1}}(undef, len(buf, step_size))
mem_spec = Array{Complex{Float64},2}(undef, len(buf, step_size), nr_taps_warped)

# get update matrices as defined in thesis
W, U, S = allpass_update_matrix_segment(nr_taps_warped, buffer_size, z_pole)

# loop through time-domain signal
for k = 1:len(buf, step_size)
    
    # update memory vector of taps
    Y = W*Y + U*buf.buffer_s
    
    # fetch taps
    mem_taps[k] = simplify(S*Y)

    # calculate windowed FFT of taps
    mem_spec[k,:] = FFTW.fft(window(nr_taps_warped).*mem_taps[k])
    
    # update buffer
    FLAG_data_remaining = step!(buf, step_size)
    
end

# discard frequency components concerning non-positive frequencies (are complex conjugate, because of real temporal signal)
if nr_taps_warped%2 == 0
    mem_spec = hcat(mem_spec[:, 2:Int(nr_taps_warped/2)])
else
    mem_spec = mem_spec[:, 2:Int(ceil(nr_taps_warped/2))]
end

# calculate real and imaginary parts of remaining frequency components
mem_spec_real = real(mem_spec)
mem_spec_imag = imag(mem_spec)
mem_spec_all = hcat(mem_spec_real, mem_spec_imag)
mem_spec_mag = log10.((mem_spec_all[:,1:length(freq)].^2 + mem_spec_all[:,length(freq)+1:end].^2))

# output signal
filter_output = [mem_taps[k][end] for k = 1:length(mem_taps)]

# plot warped spectrogram
plt.imshow(mem_spec_mag', aspect="auto", cmap="jet", origin="lower", extent=[t[1], t[end], 1, length(freq)]), plt.xlabel("time [sec]"), plt.ylabel("frequency bin"), plt.title("warped frequency power distribution")
plt.colorbar();
```

## Specify generative model

```julia
# create factor graph
fg = FactorGraph()

# placeholders
z_min = Array{Variable,1}(undef, 2*length(freq))
Î³ = Array{Variable,1}(undef, 2*length(freq))
Î¸ = Array{Variable,1}(undef, 2*length(freq))
z = Array{Variable,1}(undef, 2*length(freq))

# specify observation noise
@RV v ~ GaussianMeanVariance(zeros(buffer_size), Ïƒ_observation^2*Ic(buffer_size))

# specify previous hidden state (spectral component buffer)
for k = 1:2*length(freq)
    @RV [id=pad(:z_min,k)] z_min[k] ~ GaussianMeanVariance(placeholder(pad(:Î¼_z_min, k), dims = (AR_order,)),
                                        placeholder(pad(:Î£_z_min, k), dims = (AR_order, AR_order)))
    @RV [id=pad(:Î³,k)] Î³[k] ~ ForneyLab.Gamma(placeholder(pad(:a_Î³,k)), placeholder(pad(:b_Î³,k)))
    @RV [id=pad(:Î¸,k)] Î¸[k] ~ GaussianMeanVariance(placeholder(pad(:Î¼_Î¸,k), dims=(AR_order,)), placeholder(pad(:Î£_Î¸,k), dims=(AR_order, AR_order)))
    @RV [id=pad(:z,k)] z[k] ~ LAR.AR.Autoregressive(Î¸[k], z_min[k], Î³[k])
end

# specify Fourier matrix
eme1 = Array{Variable,1}(undef, 2*length(freq))
for k=1:2*length(freq)
    @RV eme1[k] 
    placeholder(eme1[k], pad(:eme1,k), dims = (2*length(freq), AR_order))
end

# specify previous hidden state (warped filter bank)
@RV Y_min ~ GaussianMeanVariance(placeholder(:Î¼_Y_min, dims = (2*nr_taps_warped,)),
                                 placeholder(:Î£_Y_min, dims = (2*nr_taps_warped, 2*nr_taps_warped)))

# specify inverse Fourier matrix
@RV iCS

# specify true observations
@RV x 

# update rule for hidden states
@RV Y = W*Y_min + U*x

@RV s = sum(eme1.*z)

# combine S and C to prevent posdef error :) (if order changes, ie. Y = iCs, Y will have duplicate entries due to inverse of S)
Multiplication(s, Y, iCS)
# y = Cs
# y = C*sum(em*e1*z)
# y = sum(C*em*e1*z)
# y = SY

# specify observations
@RV y = x + v

# create placeholder for fourier matrix
placeholder(iCS, :iCS, dims = (2*length(freq), nr_taps_warped))

# create placeholder for observations
placeholder(y, :y, dims = (buffer_size,))

# draw factor graph
ForneyLab.draw(fg)
ForneyLab.draw(fg, external_viewer=:default)
;
```

```julia
# specify recognition factorization and create algorithm
q = RecognitionFactorization(Î¸, z, z_min, Î³, ids=[:Î¸ :z :z_min :Î³])
algo = compatibility_fix(variationalAlgorithm(q))
# algo_F = freeEnergyAlgorithm(q) # throws error because the variables z are all connected

# evaluate algorithm
eval(Meta.parse(algo))
;
```

## Perform inference

```julia
messages=Array{Message}(undef, 190)
data = Dict()
marginals = Dict()

# create placeholders for priors
Î¼_z_min = Array{Array{Float64,1},1}(undef, 2*length(freq))
Î£_z_min = Array{Array{Float64,2},1}(undef, 2*length(freq))
a_Î³ = Array{Float64,1}(undef, 2*length(freq))
b_Î³ = Array{Float64,1}(undef, 2*length(freq))
Î¼_Î¸ = Array{Array{Float64,1},1}(undef, 2*length(freq))
Î£_Î¸ = Array{Array{Float64,2},1}(undef, 2*length(freq))


# set priors 
for k = 1:2*length(freq)
    Î¼_z_min[k] = randn(AR_order)
    Î£_z_min[k] = 1e10*Ic(AR_order)
    a_Î³[k] = 0.001
    b_Î³[k] = 0.001
    Î¼_Î¸[k] = randn(AR_order)
    Î£_Î¸[k] = 1e-6*Ic(AR_order)
end
Î¼_Y_min = randn(2*nr_taps_warped)
Î£_Y_min = 1e10*Ic(2*nr_taps_warped)

# create marginals
for k = 1:2*length(freq)
    marginals[pad(:Î³, k)] = ProbabilityDistribution(ForneyLab.Univariate, ForneyLab.Gamma, a=a_Î³[k], b=b_Î³[k])
    marginals[pad(:Î¸, k)] = ProbabilityDistribution(ForneyLab.Multivariate, GaussianMeanVariance, m=Î¼_Î¸[k], v=Î£_Î¸[k])
    marginals[pad(:z_min, k)] = ProbabilityDistribution(ForneyLab.Multivariate, GaussianMeanVariance, m=Î¼_z_min[k], v=Î£_z_min[k])
end
marginals[:Y] = ProbabilityDistribution(ForneyLab.Multivariate, GaussianMeanVariance, m=Î¼_Y_min, v=Î£_Y_min)
marginals[:Y_min] = ProbabilityDistribution(ForneyLab.Multivariate, GaussianMeanVariance, m=Î¼_Y_min, v=Î£_Y_min)

# initialize buffer
buf = BufferData(x_time, t, buffer_size)

# initialize placeholder for hidden states and observation
Î¼_x = Array{Float64,1}[]
Î£_x = Array{Float64,2}[]
Î¼_s = Array{Float64,1}[]
Î£_s = Array{Float64,2}[]

# create progress bar
p = Progress(len(buf, step_size), 1, "Observed ")

# loop through samples
for k = 1:len(buf, step_size)
        
    # update progress bar
    update!(p, k)
    
    # fill data dictionary
    data[:y] = buf.buffer_s
    for f = 1:2*length(freq)
        data[pad(:eme1,f)] = em(2*length(freq), f)*(em(AR_order, 1)')
    end
    data[:iCS] = pinv(calc_C(freq, [buf.buffer_t[1]-k/fs_desired for k=0:nr_taps_warped-1]))*S

    # infer parameters
    for it = 1:iterations
        
        # update data
        for f = 1:2*length(freq)
            data[pad(:a_Î³, f)] = a_Î³[f]
            data[pad(:b_Î³, f)] = b_Î³[f]
            data[pad(:Î¼_Î¸, f)] = Î¼_Î¸[f]
            data[pad(:Î£_Î¸, f)] = Î£_Î¸[f]
            data[pad(:Î¼_z_min, f)] = Î¼_z_min[f]
            data[pad(:Î£_z_min, f)] = Î£_z_min[f]
        end
        data[:Î¼_Y_min] = Î¼_Y_min
        data[:Î£_Y_min] = Î£_Y_min
        
        # perform steps (z last to get prediction of x)
        marginals = Base.invokelatest(stepz!, data, marginals, messages)  
        marginals = Base.invokelatest(stepz_min!, data, marginals, messages)
        marginals = Base.invokelatest(stepÎ³!, data, marginals, messages)
        marginals = Base.invokelatest(stepÎ¸!, data, marginals, messages)

        # update hidden states
        for f = 1:2*length(freq)
            Î¼_z_min[f] = ForneyLab.unsafeMean(marginals[pad(:z_min,f)])
            Î£_z_min[f] = ForneyLab.unsafeCov(marginals[pad(:z_min,f)])
            a_Î³[f] = marginals[pad(:Î³,f)].params[:a]
            b_Î³[f] = marginals[pad(:Î³,f)].params[:b]
            Î¼_Î¸[f] = ForneyLab.unsafeMean(marginals[pad(:Î¸,f)])
            #Î£_Î¸[f] = ForneyLab.unsafeCov(marginals[pad(:Î¸,f)])
        end
        Î¼_Y_min = ForneyLab.unsafeMean(marginals[:Y_min])
        Î£_Y_min = ForneyLab.unsafeCov(marginals[:Y_min])
        
    end
    
    # run another time for prediction of x
    marginals = Base.invokelatest(stepz!, data, marginals, messages)  

    # estimate prediction of x
    push!(Î¼_x, ForneyLab.unsafeMean(messages[187].dist))
    push!(Î£_x, ForneyLab.unsafeCov(messages[187].dist))
    
    # get spectral component values
    push!(Î¼_s, [ForneyLab.unsafeMean(marginals[pad(:z, k)])[1] for k=1:2*length(freq)])
    push!(Î£_s, diagm([ForneyLab.unsafeCov(marginals[pad(:z, k)])[1,1] for k=1:2*length(freq)]))
    
    # update hidden states
    for f = 1:2*length(freq)
        Î¼_z_min[f] = ForneyLab.unsafeMean(marginals[pad(:z,f)])
        Î£_z_min[f] = ForneyLab.unsafeCov(marginals[pad(:z,f)])
        a_Î³[f] = marginals[pad(:Î³,f)].params[:a]
        b_Î³[f] = marginals[pad(:Î³,f)].params[:b]
        Î¼_Î¸[f] = ForneyLab.unsafeMean(marginals[pad(:Î¸,f)])
        #Î£_Î¸[f] = ForneyLab.unsafeCov(marginals[pad(:Î¸,f)])
    end
    Î¼_Y_min = ForneyLab.unsafeMean(marginals[:Y])
    Î£_Y_min = ForneyLab.unsafeCov(marginals[:Y])
    
    # update buffer
    FLAG_data_remaining = step!(buf, step_size)
    
end
```
